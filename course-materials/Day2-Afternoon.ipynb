## 2. Building blocks and pathways

This morning's exercise was intended to highlight the way that different notebooks, tools, datasets, and processes might be connected together to build research workflows – ways of focusing in on particular questions.

As you've been working through all the different apps and notebooks in the GLAM Workbench over the last couple of days you've probably also started to notice a few recurring patterns and themes. Things like:

* downloading files with `requests`
* opening, saving, and analysing CSV files with Pandas
* making charts with Altair
* constructing loops to harvest data by making multiple requests to an API

I find myself repeating these sorts of patterns over and over again. As I said on the first day, recognising and reflecting on these sorts of patterns can help you design your own processing pipelines for GLAM data.

This afternoon we're going to think about how we might develop a new notebook (or set of notebooks) to achieve particular tasks. Here's some possibilities:

* With the demise of the Australian Women's Weekly portal in the recent Trove redesign, we've lost the ability to browse all those funky covers. Could we build some sort of replacement? There's a notebook that [downloads page images](../trove-newspapers/Save-page-image.ipynb), and one that gets [front pages from 100 years ago](../trove-newspapers/Todays-news-yesterday.ipynb), could we connect them up in some way?
* The State Library of Victoria doesn't have a public API, but it does make lots of images available to [download via the IIIF protocol](../state-library-vic/download_image_from_iiif.ipynb). SLV records are also in Trove. Is there some way of harvesting SLV images from IIIF via a search in Trove? And can we make use of [some cool IIIF tricks](../state-library-vic/more_fun_with_iiif.ipynb)?
* Robyn raised a question about extracting data from the [NT Parliamentary record](https://www.territorystories.nt.gov.au/jspui/handle/10070/220000#) (as we have from Commonwealth Hansard). I really don't know what's possible, but it might be worth a look.

We'll be looking at one of more of these, trying to break them down into their components.

## 3. Activity – Designing your own project

As I said at the start of the workshop, I'm hoping that you will all be able to make something by the end of tomorrow. It may be that you want to go back to one of the tools we've already looked at and modify it to use a different dataset. Do you have a collection of CSV files you'd like to load into your own version of the GLAM CSV Explorer? Or perhaps you want to go back up to this morning's exercise and do a **big** harvest of newspaper articles to analyse or share.

This afternoon I want you to sketch out your project. As we did together in the previous section, break your idea down into a series of resources and steps.

1. What data do you want to use?
2. How can you access the data?
3. What processing steps will be required to get the data in the form you need.
4. Are there any existing notebooks in the GLAM Workbench which do something similar? How could they be modified?
4. What would the end product look like?

Not all the projects might be do-able in the time we have. But I'm sure we'll be able to make a useful start.

To help you in your planning, I've created a list of notebooks based not on the source of the data, but on the types of data and processes they use. Mix and match to sketch out your own workflow!

And of course, don't forget to include the [lists of GLAM data](https://glam-workbench.github.io/glam-data-list/) we looked at yesterday!
